```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

# defino de la funcion gaussiana
def gaussian(x, mu=0, sigma=1):
    return np.exp(-((x - mu)**2) / (2 * sigma**2))

# puntos de interpolación
x_points = np.array([0, 1, 2])
y_points = gaussian(x_points)

# construccion del polinomio interpolante de lagrange
polynomial = lagrange(x_points, y_points)

# Valores de la gaussiana y del polinomio interpolante para comparación
x_vals = np.linspace(0, 4, 400)
gaussian_vals = gaussian(x_vals)
polynomial_vals = polynomial(x_vals)

# Evaluación en x = 3
x_eval = 3
gaussian_eval = gaussian(x_eval)
polynomial_eval = polynomial(x_eval)

# Gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_vals, gaussian_vals, label='Gaussiana', color='blue')
plt.plot(x_vals, polynomial_vals, '--', label='Polinomio interpolante', color='red')
plt.scatter(x_points, y_points, color='black', zorder=5)
plt.axvline(x=x_eval, color='green', linestyle=':', label=f'x = {x_eval}')
plt.scatter(x_eval, gaussian_eval, color='blue', zorder=6)
plt.scatter(x_eval, polynomial_eval, color='red', zorder=6)

plt.title('Aproximación polinómica de la funcion gaussiana')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()

# Resultados numéricos
print(f'Valor de la gaussiana en x = {x_eval}: {gaussian_eval}')
print(f'Valor del polinomio interpolante en x = {x_eval}: {polynomial_eval}')

#acá se observa el grafico que compara la funcion gaussiana con su aproximacion polinomica
# usando la interpolacion de lagrange en el intervalo [0,2]
# se ve que el polinomio se desvia del valor real de la gaussiana
# esto muestra que la extrapolacionh fuera del intervalo no es precisa
# por lo tanto la respiuesta es no, no puede aproximar bien la funcion gaussiana
```
