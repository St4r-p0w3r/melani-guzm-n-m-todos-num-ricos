Escriba el polinomio de Mclaurin de sexto orden para la función  xex  y use los polinomios de Chebyshev para encontrar el polinomio de el orden menor posible que mantenga el error menor a  0.01  sobre el intervalo  [−1,1] .

Pista: Recuerde que el valor máximo de  Tn  sobre  [−1,1]  es  1 . Use la función chebval de numpy.polynomial.chebyshev

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial.chebyshev import Chebyshev, chebval

# defino la función f(x) = x * exp(x)
def f(x):
    return x * np.exp(x)

# puntos de evaluacion en el intervalo [-1, 1]
x_vals = np.linspace(-1, 1, 400)
y_vals = f(x_vals)

# inicializamos el orden del polinomio de chebyshev
max_error = 0.01
n = 1  # empezamos con un polinomio de Chebyshev de primer orden
error = float('inf')


# busco el orden minimo tal que el error sea menor que 0.01
while error > max_error:
    # se aproxima la función f(x) con un polinomio de chebyshev de orden n
    c = Chebyshev.fit(x_vals, y_vals, n)
    
    # evaluamos el polinomio de chebyshev en los puntos x_vals
    approx_vals = chebval(x_vals, c.coef)
    
    # se calcula el error máximo
    error = np.max(np.abs(approx_vals - y_vals))
    
    #iIncremento el orden
    n += 1

# muestro el orden del polinomio y el error alcanzado
print(f"Orden mínimo del polinomio de Chebyshev: {n - 1}")
print(f"Error máximo alcanzado: {error}")

# se grafica la funcion original y la aproximacion con el polinomio de chebyshev
plt.plot(x_vals, y_vals, label='f(x) = x * exp(x)', color='blue')
plt.plot(x_vals, approx_vals, label=f'polinomio de chebyshev (Orden {n - 1})', color='red', linestyle='dashed')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('aproximacion de f(x) con Polinomio de chebyshev')
plt.grid(True)
plt.show()
```
